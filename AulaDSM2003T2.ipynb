```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import requests
```

# Carregamento da API Alpha Vantage


```python
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=IBM&outputsize=full&apikey=NES1FN2QJITLR295'

r = requests.get(url)
data = r.json()

time_series = data["Time Series (Daily)"]

base_acoes = pd.DataFrame.from_dict(time_series, orient='index')

base_acoes.columns  = ['Abertura', 'Alta', 'Baixa', 'Fechamento', 'Volume']
base_acoes = base_acoes.astype(float)
base_acoes.index = pd.to_datetime(base_acoes.index)
base_acoes = base_acoes.sort_index()
```

# Engenharia de Classificadores Categoricos para Aprendizagem


```python
base_acoes.describe()
```

```python
base_acoes["D_Fechamento"] = base_acoes["Fechamento"].diff().shift(-1)
base_acoes["D_Volume"] = base_acoes["Volume"].diff().shift(-1)

def classifica_tendencia(row):
  if row["D_Fechamento"] > 0:
      return "Alta com força" if row["D_Volume"]> 0 else "Distribuição"
  else:
      return "Sem convicção" if row["D_Volume"] > 0 else "Sinal de exaustão"

base_acoes["Momento de: "] = base_acoes.apply(classifica_tendencia, axis=1)
base_acoes
```
